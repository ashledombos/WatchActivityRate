# Exploration de la base de données SQLite avec Pandas

```python
import sqlite3
import pandas as pd
```

## Connexion à la base de données SQLite

```python
conn = sqlite3.connect('activity_data.db')  # Remplacez par le chemin de votre base de données SQLite
```

## Lecture des données dans un DataFrame Pandas

```python
df = pd.read_sql_query('SELECT * FROM activity', conn)  # Remplacez 'activity' par le nom de votre table
```
